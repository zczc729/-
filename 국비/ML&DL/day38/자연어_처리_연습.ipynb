```python
korean_stopwords = korean_stopwords + not_in_stopwords
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

word_tokens = word_tokenize(meeting_text)

result = []
for w in word_tokens: 
    if w not in korean_stopwords: 
        result.append(w) 

print(result)
```

```python
result
```

```python
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
```

```python
ko = nltk.Text(result)
ko
```

```python
ko.collocations()
```

```python
ko.vocab()
```

```python
ko_most = ko.vocab().most_common(500)
ko_most
dict(ko_most)
```

```python
# WordCloud 이미지 생성:
wordcloud = WordCloud(font_path='/System/Library/Fonts/Supplemental/AppleSDGothicNeo.ttc',
                      background_color='white', width=1920, height=1080).generate_from_frequencies(dict(ko_most))
```

```python
# 화면에 출력:
plt.figure(figsize=(15,9))
plt.imshow(wordcloud, interpolation='bilinear', cmap='YlOrBr')
plt.savefig('./img/wordcloud.png', format='png', dpi=300, bbox_inches='tight')
plt.axis('off')
plt.show()
```
