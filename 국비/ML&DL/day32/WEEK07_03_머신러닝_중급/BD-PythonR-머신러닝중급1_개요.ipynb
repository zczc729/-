#### iris dataset 분리 작업: train 데이터와 test 데이터 분리

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, 
                                                    test_size=0.2,          # train:test = 80:20
                                                    random_state=1)         # 예컨대, seed번호
```

#### train 데이터로 학습(Fit) 수행 :

```python
# 분류알고리즘 객체 생성:
dt_clf = DecisionTreeClassifier(random_state=9)
```

```python
# 학습 수행:
dt_clf.fit(X_train, y_train)
```

#### test 데이터로 예측(Predict) 수행 :

```python
# 학습이 완료된 객체로부터 Test 데이터 예측 수행:
y_predict = dt_clf.predict(X_test)
```

```python
y_predict
```

#### 예측 정확도 평가(Evaluation) :

```python
pd.DataFrame({"y_test": y_test, "predict": y_predict, "Accuracy": y_test == y_predict})
```

```python
from sklearn.metrics import accuracy_score
#accuracy_score(y_test, predict)
print(f'Accuracy: {accuracy_score(y_test, y_predict)}')
```

# 실습)

```python
for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=i)         # 예컨대, seed번호
    # for j in range(11):
    dt_clf = DecisionTreeClassifier(random_state=9)

    dt_clf.fit(X_train, y_train)
    y_predict = dt_clf.predict(X_test)
    print(f'{i}번째 Accuracy: {accuracy_score(y_test, y_predict)}')
```
