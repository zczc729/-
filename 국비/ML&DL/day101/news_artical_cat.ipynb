```python
import requests
from bs4 import BeautifulSoup

# 요청할 URL 설정
url = "https://www.naeil.com/news/read/526560"

# HTTP GET 요청 보내기
response = requests.get(url, headers={'User-agent': 'Mozilla/5.0'})
response.encoding = 'utf-8'
# 상태 코드가 200일 때만 진행
if response.status_code == 200:
    # HTML 파싱
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    # class가 'art_body'인 첫 번째 요소 찾기
    art_body = soup.find(class_="article-view")
    
    # 추출한 내용의 텍스트만 출력
    if art_body:
        print(art_body.get_text(strip=True, separator='\n'))
    else:
        print("class='art_body'를 찾을 수 없습니다.")

else:
    print(f"요청 실패: 상태 코드 {response.status_code}")
```
