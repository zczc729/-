```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE

# 데이터 준비
X = train_df_f.drop(columns='우울증 현재 유병 여부')
y = train_df_f['우울증 현재 유병 여부']

# 다중 클래스 레이블 인코딩
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 데이터셋을 트레인과 테스트로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SMOTE 적용
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

# 단계적 선택법 구현
def stepwise_selection(X_train, y_train):
    initial_features = []
    best_features = []
    current_score = 0.0
    best_score = 0.0
    
    while True:
        remaining_features = list(set(X_train.columns) - set(best_features))
        scores_with_candidates = []
        
        # 전진 선택: 남아 있는 변수 중에서 가장 좋은 것을 추가
        for feature in remaining_features:
            current_features = best_features + [feature]
            model = LogisticRegression(max_iter=1000)
            scores = cross_val_score(model, X_train[current_features], y_train, cv=5)
            score = np.mean(scores)
            scores_with_candidates.append((score, feature))
        
        # 후진 소거: 기존의 변수 중에서 가장 성능이 떨어지는 것을 제거
        for feature in best_features:
            current_features = list(set(best_features) - set([feature]))
            if len(current_features) == 0:
                continue
            model = LogisticRegression(max_iter=1000)
            scores = cross_val_score(model, X_train[current_features], y_train, cv=5)
            score = np.mean(scores)
            scores_with_candidates.append((score, feature))
        
        scores_with_candidates.sort(reverse=True)
        best_new_score, best_feature = scores_with_candidates[0]
        
        if best_new_score > current_score:
            current_score = best_new_score
            if best_feature in remaining_features:
                best_features.append(best_feature)
            else:
                best_features.remove(best_feature)
        else:
            break
    
    return best_features

X_train_df = pd.DataFrame(X_train_smote, columns=X.columns)
best_features = stepwise_selection(X_train_df, y_train_smote)

# 선택된 특성으로 로지스틱 회귀 모델 생성 및 훈련
model = LogisticRegression(max_iter=1000)
model.fit(X_train_df[best_features], y_train_smote)

# 모델 평가 및 예측
X_test_df = pd.DataFrame(X_test_scaled, columns=X.columns)
y_pred = model.predict(X_test_df[best_features])
y_pred_prob = model.predict_proba(X_test_df[best_features])  # 각 클래스에 대한 예측 확률

# 성능 평가
report = classification_report(y_test, y_pred)
print(report)

# 혼동 행렬 시각화
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# 특성 중요도 시각화
feature_importance = np.abs(model.coef_[0])
selected_feature_names = best_features

# 특성 중요도와 특성 이름을 DataFrame으로 만들기
importance_df = pd.DataFrame({'Feature': selected_feature_names, 'Importance': feature_importance})

# 중요도 순으로 정렬
importance_df = importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance in Logistic Regression')
plt.show()
```
