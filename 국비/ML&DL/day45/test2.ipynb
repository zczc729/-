```python
import pandas as pd
import numpy as np
```

```python
pill = pd.read_csv('./data/약물_전체.csv')
```

```python
np.NAN
```

```python
atc_01 = []
atc_02 = []
atc_03 = []
atc_04 = []
atc_05 = []
def atc_sep(row):
    if pd.isnull(row):
        atc_01.append(np.NAN)
        atc_02.append(np.NAN)
        atc_03.append(np.NAN)
        atc_04.append(np.NAN)
        atc_05.append(np.NAN)
    else:
        atc_len = len(row)
        if atc_len <= 3:
            atc_01.append(row[0])
            atc_02.append(row[1:atc_len])
            atc_03.append(np.NAN)
            atc_04.append(np.NAN)
            atc_05.append(np.NAN)
        elif atc_len == 4:
            atc_01.append(row[0])
            atc_02.append(row[1:3])
            atc_03.append(row[3])
            atc_04.append(np.NAN)
            atc_05.append(np.NAN)
        elif atc_len == 5:
            atc_01.append(row[0])
            atc_02.append(row[1:3])
            atc_03.append(row[3])
            atc_04.append(row[4])
            atc_05.append(np.NAN)
        else:
            atc_01.append(row[0])
            atc_02.append(row[1:3])
            atc_03.append(row[3])
            atc_04.append(row[4])
            atc_05.append(row[5:])

            
```

```python
pill['ATC코드'].apply(atc_sep)
```

```python
pill['ATC_01'] = atc_01
pill['ATC_02'] = atc_02
pill['ATC_03'] = atc_03
pill['ATC_04'] = atc_04
pill['ATC_05'] = atc_05
```

```python
pill.to_csv("./data/atc코드.csv")
```
