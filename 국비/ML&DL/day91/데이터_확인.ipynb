```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# 1. MPS 장치가 있는지 확인
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")

# 2. 토크나이저와 모델 불러오기
tokenizer = AutoTokenizer.from_pretrained("WhitePeak/bert-base-cased-Korean-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("WhitePeak/bert-base-cased-Korean-sentiment")

# 3. 모델을 MPS 장치로 이동
model.to(device)

# 4. 분석할 문장 정의
text = df['sentence_form'] # 감정 분석할 텍스트

# 5. 입력 데이터를 토큰화
for t in text:
    inputs = tokenizer(t, return_tensors="pt")  # PyTorch tensor로 변환

    # 6. 입력 데이터를 MPS 장치로 이동
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # 7. 모델로 예측 수행
    with torch.no_grad():
        outputs = model(**inputs)

    # 8. 결과 해석 (로그 확률을 소프트맥스하여 감정 점수로 변환)
    probs = torch.softmax(outputs.logits, dim=-1)

    # 9. 긍정/부정 결과 도출
    predicted_class = torch.argmax(probs).item()  # 가장 높은 점수의 클래스 선택
    print(predicted_class)
    labels = ['negative', 'positive', 'neutral']  # 모델에 따라 달라질 수 있는 레이블
    predicted_label = labels[predicted_class]

    # 결과 출력
    # print(f"입력 텍스트: {text}")
    print(f"예측된 감정: {predicted_label}")
```

```python
from transformers import pipeline

sentiment_analyzer = pipeline("sentiment-analysis", model="WhitePeak/bert-base-cased-Korean-sentiment")

def analyze_sentiment(text, neutral_threshold=0.2):
    result = sentiment_analyzer(text)[0]
    positive_score = result['score'] if result['label'] == 'LABEL_1' else 1 - result['score']
    negative_score = 1 - positive_score
    
    if abs(positive_score - negative_score) < neutral_threshold:
        return "중립"
    elif positive_score > negative_score:
        return "긍정"
    else:
        return "부정"

# 사용 예시
text = "이 영화는 그저 그랬어요."
sentiment = analyze_sentiment(text)
print(f"감성 분석 결과: {sentiment}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# 모델과 토크나이저 로드
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")

model_name = "WhitePeak/bert-base-cased-Korean-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.to(device)


def analyze_sentiment(text, neutral_threshold=0.2):
    # 입력 텍스트 토큰화
    inputs = tokenizer(text, return_tensors="pt")
    
    # 모델 추론
    with torch.no_grad():
        outputs = model(**inputs)
    
    # 소프트맥스를 통해 확률 계산
    probabilities = F.softmax(outputs.logits, dim=-1)
    
    # 긍정 확률 (LABEL_1에 해당)
    positive_score = probabilities[0][1].item()
    negative_score = probabilities[0][0].item()
    
    # 감성 판단
    if abs(positive_score - negative_score) < neutral_threshold:
        return "neutral"
    elif positive_score > negative_score:
        return "positive"
    else:
        return "negative"

# 사용 예시
text = df['sentence_form']
result = []
for i in range(len(text)):
    sentiment = analyze_sentiment(text[i])
    # print(f"감성 분석 결과: {sentiment}")
    print(i)
    result.append(sentiment)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# MPS 디바이스 설정
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
print(f"Using device: {device}")

# 모델과 토크나이저 로드
model_name = "WhitePeak/bert-base-cased-Korean-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.to(device)

def process_batch(batch_texts, batch_size=512):
    batch_results = []
    for i in range(0, len(batch_texts), batch_size):
        batch = batch_texts[i:i+batch_size]
        # 각 텍스트를 문자열로 변환
        batch = [str(text) for text in batch]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}
        
        with torch.no_grad():
            outputs = model(**inputs)
        
        probabilities = F.softmax(outputs.logits, dim=-1)
        
        for prob in probabilities:
            positive_score = prob[1].item()
            negative_score = prob[0].item()
            
            if abs(positive_score - negative_score) < 0.1:
                batch_results.append("neutral")
            elif positive_score > negative_score:
                batch_results.append("positive")
            else:
                batch_results.append("negative")
        
        print(f"Processed {i+len(batch)} out of {len(batch_texts)}")
    
    return batch_results

# DataFrame의 'sentence_form' 열을 리스트로 변환
text_list = df['sentence_form'].tolist()

# 배치 처리 실행
result = process_batch(text_list)

print("Analysis complete!")

df['model_result'] = result


test_texts = df['p_n']
true_labels = df['model_result']


# 평가 지표 계산
accuracy = accuracy_score(true_labels, test_texts)
precision = precision_score(true_labels, test_texts, average='weighted')
recall = recall_score(true_labels, test_texts, average='weighted')
f1 = f1_score(true_labels, test_texts, average='weighted')

# 결과 출력
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
