## 텍스트 마이닝 개요



#### 텍스트 마이닝 의미 :

### 토큰 생성 (Tokenizing) 

 - 개념 정의 : 자연어 문서 집합을 위한, 작은 단위로 나누는 작업
 - 토큰(Token) : 나눈 작은 문자열 단위
 - 토큰화(Tokenizing) : 문자열을 토큰으로 나누는 작업 (토큰 생성)

```python
# 문장 단위의 토큰 리스트 :
from nltk.tokenize import sent_tokenize
print(sent_tokenize(emma[:1000]))
```

```python
print(sent_tokenize(emma[:1000])[3])
```

```python
# 단어 단위의 토큰 리스트 :
from nltk.tokenize import word_tokenize
print(word_tokenize(emma[:100]))
```

```python
print(word_tokenize(sent_tokenize(emma[:1000])[3]))
```

### 형태소(Morpheme) 분석 

 - 개념 정의 : 단어로부터 어근, 접두사, 접미사, 품사 등 언어적 속성을 파악하는 작업
 - 형태소(Morpheme) : 의미가 있는 가장 작은 말의 단위
 - 주요 작업 : 어간 추출(Stemming), 원형 복원(Lemmatizing), 품사 부착(Part-Of-Speech Tagging)

### Text 클래스 

 - 텍스트 문서 분석에 유용한 메소드 제공

```python
# 정규표현식(RegexpTokenizer) 을 이용해 Text 객체로 만들기
import nltk
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
emma
```

```python
from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")        
# \w 는 word 단어군을 의미(알파벳, 숫자, _),  + 는 1번 이상 반복됨을 의미
retokenize
```

```python
from nltk import Text
emma_text = Text(retokenize.tokenize(emma), name='Emma')
emma_text
```

```python
# plot() 메소드를 사용해 단어(Token) 사용 빈도를 그래프로 표현
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 3))
emma_text.plot(20)
```

```python
# 해당 단어가 들어간 문장을 찾아봅니다 
emma_text.concordance('have', lines=10)
```

```python
# 비슷한 단어 찾아봅니다 
emma_text.similar('general')
```

```python
emma_text.similar('general', 10)
```

```python
emma_text.similar('strong')
```

```python
# 여러 개 단어에 의해 공유되는 문맥을 조사합니다
emma_text.common_contexts(['general','strong'])
```

```python
# 전체 텍스트 시각화 
plt.figure(figsize=(18, 5))
emma_text.dispersion_plot(['Emma','Kightley','Frank','Robert','Harriet'])
```

```python
# 단어의 총수를 계산 :
len(emma_text)
```

```python
# 중복을 제거한 단어의 수 :
len(set(emma_text))
```

```python
# 단어의 풍부성 측정 :
len(set(emma_text)) / len(emma_text)
```

#### 1) 꼬꼬마 엔진을 통한 한글분석
- developed by the Intelligent Data Systems (IDS) Laboratory at SNU



```python
from konlpy.tag import Kkma
kkma = Kkma()
kkma
```

```python
# 문장(Sentence) :
kkma.sentences('2024년 천연가스 가격에 대한 전망은 여러 요인에 따라 다소 복잡하게 전개되고 있습니다.')
```

```python
sample = '2024년 천연가스 가격에 대한 전망은 여러 요인에 따라 다소 복잡하게 전개되고 있습니다 한국어 분석'
kkma.sentences(sample)   # 마침표가 없어도 문장(Sentence) 로 구분
```

```python
# 형태소 분석 :
kkma.morphs(sample)
```

```python
# 품사; 명사(Noun) :
kkma.nouns(sample)
```

```python
# 구문(Phrase) :
kkma.pos(sample)   # NNG(명사), OL(외국어), VCP(조사), EFN(종결어미)
```

### 한글 분석 : 시각화 연습
- KoNLPy 내장 법률문서 활용

```python
import nltk
```

```python
from konlpy.corpus import kolaw     # kolaw : 대한민국 법률 말뭉치
law_ko = kolaw.fileids()
law_ko
```

```python
from konlpy.corpus import kobill    # kobill: 대한민국 국회입법안 말뭉치
bill_ko = kobill.fileids()
bill_ko
```

```python
doc_law_ko = kolaw.open('constitution.txt').read()
print(doc_law_ko)
```

```python
doc_bill_ko = kobill.open('1809898.txt').read()
print(doc_bill_ko)
```

```python
# from konlpy.tag import Twitter
# twit = Twitter()
# twit
from konlpy.tag import Okt
okt = Okt()
okt
```

```python
# token_ko = twit.nouns(doc_law_ko)     # 명사 분석
# token_ko
token_ko = okt.nouns(doc_law_ko)     # 명사 분석
token_ko
```

```python
ko = nltk.Text(token_ko, name='법률')
ko
```

```python
print(len(ko.tokens))       # token 갯수; '법률' 글자가 들어간 문장이 몇개 들어있는지
```

```python
print(len(set(ko.tokens)))  # 중복 제외한 token 갯수
```

```python
### 아래 에러 발생시
#import nltk
#nltk.download('stopwords')
```

```python
ko.vocab()                  # token 빈출정도 정규분포
```

```python
import matplotlib
matplotlib.rc('font', family='Apple SD Gothic Neo') # for Windows 한글 깨지는 문제 해결
```

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
ko.plot(50)
plt.show()
```

```python
# 전체 텍스트 시각화 
plt.figure(figsize=(18, 5))
ko.dispersion_plot(['대통령','국회의원','공무원'])
```

```python
# WordCloud 시각화
ko.collocations()
```

```python
ko.vocab()
ko_most = ko.vocab().most_common(500)
ko_most
```

```python
dict(ko_most)
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=alice_mask)
wc = wc.generate_from_frequencies(dict(ko_most)) 
wc
# Create a word_cloud from words and frequencies
```

```python
# 한글 폰트 깨지는 문제 해결 :
wc = WordCloud(font_path='/System/Library/Fonts/Supplemental/AppleSDGothicNeo.ttc',
               background_color="white", max_words=1000, mask=alice_mask).generate_from_frequencies(dict(ko_most))
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
