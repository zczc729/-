```python
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Text 파일 열기:
text = open('data/message_kr.txt', encoding="utf8").read()
text
```

```python
# 한글 형태소 분석:
import nltk
from konlpy.tag import Okt
okt = Okt()
okt
```

```python
token_ko = okt.nouns(text)     # 명사 분석
token_ko
```

```python
ko = nltk.Text(token_ko, name='법률')
ko
```

```python
ko.collocations()
```

```python
ko.vocab()
```

```python
ko_most = ko.vocab().most_common(500)
ko_most
dict(ko_most)
```

```python
# WordCloud 이미지 생성:
wordcloud = WordCloud(font_path='/System/Library/Fonts/Supplemental/AppleSDGothicNeo.ttc',
                      background_color='white', width=1920, height=1080).generate_from_frequencies(dict(ko_most))
```

```python
# 화면에 출력:
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear', cmap='YlOrBr')
plt.axis('off')
plt.show()
```

## 실습 (2) 포털 사이트 탐색결과 텍스트 분석

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from urllib.request import urlopen
```

#### 1) 포털 사이트로부터 검색어 결과 수집

```python
# 포털 검색어 결과 수집
url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%EC%BD%94%EB%A1%9C%EB%82%98'
response = urlopen(url)
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
soup
```

```python
tmp = soup.find_all('a')
tmp
```

```python
list_tmp = []
for line in tmp:
    list_tmp.append(line.text)
    
list_tmp
```

#### 2) 텍스트 분석 및 시각화

```python
import nltk
from konlpy.tag import Okt
okt = Okt()
okt
```

```python
# 하나의 글로 만드는 작업 :
text_tmp = ''

for each_line in list_tmp:
    text_tmp = text_tmp + each_line + '\n'
```

```python
text_tmp
```

```python
# 형태소 분석 :
token_ko = okt.morphs(text_tmp)
token_ko
```

```python
# 형태소에 포함된 단어 갯수 분석 :
ko = nltk.Text(token_ko, name='코로나')# '코로나' 포함된 단어
print(len(ko.tokens))                 # 분석한 단어 갯수
print(len(set(ko.tokens)))            # 분석한 단어 갯수 중복된 단어 제거 갯수
```

```python
ko.vocab()
ko.vocab().most_common(100)
```

```python
# 빈도수 그려보기 
import matplotlib
plt.figure(figsize=(15,6))
matplotlib.rc('font', family='Apple SD Gothic Neo') # for Windows 한글 깨지는 문제 해결
ko.plot(50)
plt.show()
```

```python
token_ko
```

```python
# stopwords(불용어) 처리 :

stop_words = ['.','가','`','"','0','(',',','...','에서','\n','\n\n','\xa0','\n\n\n', '\n  \n']

result = [] 
for w in token_ko: 
    if w not in stop_words: 
        result.append(w) 
# 위의 4줄은 아래의 한 줄로 대체 가능
# result = [each_word for each_word in token_ko if each_word not in stop_words]

ko = nltk.Text(result, name='코로나')
ko.vocab().most_common(50)
```

```python
# 빈도수 그려보기 
import matplotlib
matplotlib.rc('font', family='Apple SD Gothic Neo') # for Windows 한글 깨지는 문제 해결
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```

```python
# WordCloud 에 그려보기 :
ko_most = ko.vocab().most_common(50)
ko_most
```

```python
# Create a word_cloud from words and frequencies
wc = WordCloud(font_path='/System/Library/Fonts/Supplemental/AppleSDGothicNeo.ttc',
               background_color="white", max_words=1000).generate_from_frequencies(dict(ko_most))

# wc = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
#                background_color="white", max_words=1000, mask=alice_mask).generate_from_frequencies(dict(ko_most))
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
