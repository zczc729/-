### 2-4) 날짜 및 시간 세팅

```python
timezone_offset = weather_data['timezone_offset']
timezone_delta = datetime.timedelta(seconds=timezone_offset)
```

```python
timezone_delta = datetime.timedelta(seconds=timezone_offset)
timezone_delta
```

```python
for col in ['dt', 'sunrise', 'sunset']:
    current_df[col] = current_df[col].apply(lambda x: datetime.datetime.utcfromtimestamp(x) + timezone_delta)
    current_df[col] = current_df[col].dt.strftime('%Y-%m-%d %H:%M:%S')
current_df
```

```python
minutely_df['dt'] = minutely_df['dt'].apply(lambda x: datetime.datetime.utcfromtimestamp(x) + timezone_delta)
minutely_df['dt'] = minutely_df['dt'].dt.strftime('%Y-%m-%d %H:%M:%S')
```
