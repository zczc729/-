판다스 패키지에서 자주 사용하는 명령어를 간단히 설명하고 사용하는 예를 들어보시오. 

**기술 사항**   
- DataFrame의 구조 형태 및 딕셔너리에서 데이터프레임 생성 예시 (5점)    
- read_csv의 기능 및 CSV 파일 데이터 읽기 예시 (5점)    
- describe() 내장함수의 기능 및 사용 예시 (5점)    
- pivot_table() 내장함수의 기능 및 data, index, values 옵션을 사용한 예시 (5점)   

- 답변)

```python
DataFrame의 구조 형태 및 딕셔너리에서 데이터 프레임을 생성할 시 별다른 조건 없이 바로 변환이 가능하다.
따라서,

dict_test = {'A' : 1, 'B' : 2}

df = pd.DataFrame(dict_test)

이러한 형태로 바로 변환이 가능하다.
```

```python
read_csv() 기능은 csv 형태로 저장되어 있는 파일을 읽어 들어와 데이터 프레임화 시키는 것이다.
pd.read_csv('파일 경로', encoding='원하는 인코딩')

이러한 형태로 주로 쓰인다.
```

```python
describe()는 주로 해당 데이터 프레임의 평균, 표준 편차, max, min 등등 통계적 정보들이 필요할 때 쓰인다.
data.describe(include='all')

include='all' 옵션을 통해 결측값도 포함하여 출력 가능하다. 기본적으로 결측값은 출력되지 않는다.
```

```python
pivot_table()은 pivot 기능을 활용하여 테이블화 시키는데 사용된다. 열(column) 값에 따라 행과 열들을 재구성 하여 집계 하는데 사용된다.
data : 읽어 들여올 데이터
index : 인덱스로 사용할 열의 이름이나 열들의 리스트
values : 집계할 열의 이름이나 열들의 리스트
    
pd.pivot_table(data=data, index='측정일', values='몸무게')

data의 측정일을 기준으로 몸무게의 평균을 구한다.
```
