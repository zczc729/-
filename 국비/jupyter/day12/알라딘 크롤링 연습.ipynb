```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=1&cnt=1000&SortOrder=1'
res = requests.get(url)
res
```

```python
soup = BeautifulSoup(res.text, "html.parser")
soup
```

```python
#Myform > div:nth-child(3) > table > tbody > tr > td:nth-child(3) > table > tbody > tr:nth-child(1) > td:nth-child(1) > div:nth-child(1) > ul > li:nth-child(2) > a > b
book_name_soup = soup.select('li:nth-child(2) > a > b')
book_name = []

for i in range(len(book_name_soup)):
    book_name.append(book_name_soup[i].text)
    
book_name
```

```python
#Myform > div:nth-child(3) > table > tbody > tr > td:nth-child(3) > table > tbody > tr:nth-child(1) > td:nth-child(1) > div:nth-child(1) > ul > li:nth-child(3)
book_info_soup = soup.select('td:nth-child(1) > div:nth-child(1) > ul > li:nth-child(3)')
book_info = []

for i in range(len(book_info_soup)):
    ls = book_info_soup[i].text
    book_info.append(ls.split('|'))
    
book_info
```

```python
book_price_soup = soup.find_all('span', class_ = 'ss_p2')
book_price = []

for i in range(len(book_price_soup)):
    book_price.append(book_price_soup[i].text)
    
book_price
```

```python
book_url_soup = soup.find_all('a', class_ = 'bo3')
book_url = []

for i in range(len(book_url_soup)):
    book_url.append(book_url_soup[i]['href'])
    
book_url
```
