```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url_base = "https://www.chicagomag.com"
url_sub = "/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/"
url = url_base + url_sub
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
res = requests.get(url, headers=headers)
```

```python
res
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text)
soup
```

```python
sammy = soup.find_all('div', class_='sammy')
len(sammy)
sammy[0]
```

```python
sammy_rank = soup.find_all('div', class_='sammyRank')
sammy_rank[0].text
```

```python
sammyList = soup.find_all('div', class_='sammyListing')
sammyList[0].get_text()
```

```python
import pandas as pd
import re
df = pd.DataFrame()
```

```python
for i in range(len(sammy)):
    df['Rank'] = sammy[i].text
    
df['Rank'] = [sammy_rank[i].text for i in range(len(sammy))]
```

```python
re.split(('\n|\r\n'), sammyList)
```

```python
from urllib.request import urljoin
```

```python
rank = []
main_menu = []
cafe_name = []
url = []

for i in sammy:
    rank.append(i.find('div', class_ = 'sammyRank').get_text())
    menu_text = i.find('div', class_ = 'sammyListing').get_text()
    
    main_menu.append(re.split(('\n|\r\n'), menu_text)[0])
    cafe_name.append(re.split(('\n|\r\n'), menu_text)[1])
    
    url.append(urljoin(url_base, i.find('a')['href']))
    

    
```

```python
len(rank), len(main_menu), len(cafe_name), len(url)
```

```python
data = {'Rank':rank, 'Cafe':cafe_name, 'Menu':main_menu, 'URL':url}
df = pd.DataFrame(data)
# df.set_index('Rank', inplace=True)
```

```python
df.to_csv('./WEEK03_04_웹크롤링_기초/data/best_sandwiches_list_chicago_raw_my.csv', sep=",", encoding='UTF-8')
```
