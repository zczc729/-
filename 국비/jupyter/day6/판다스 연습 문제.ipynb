10. 데이터프레임의 통계적 요약 정보를 출력해보세요(평균, 표준편차, 최소값, 최대값 등).

```python
pd.DataFrame(df.mean().to_frame(name='평균'))
```

```python
pd.DataFrame(df.min().to_frame(name='최소값'))
```

```python
pd.DataFrame(df.max().to_frame(name='최대값'))
```

```python
pd.DataFrame(df.std().to_frame(name='표준편차'))
```

```python
df = df.describe().drop(['25%', '50%', '75%'])
# df.rename(index='평균값','중간값','표준편차','최소값','최대값')
df.rename(index={'count': '평균값', 'mean': '중간값', 'std': '표준편차', 'min': '최소값', 'max': '최대값'}, inplace=True)
df
```

```python
for col in df.columns:
    if df[col].dtype == float:
        df[col] = df[col].apply(lambda x:int(x) if str(x).split('.')[-1] == '0' else round(x, 2))
df
```
