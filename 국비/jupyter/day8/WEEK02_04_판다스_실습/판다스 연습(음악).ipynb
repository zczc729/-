```python
import pandas as pd
```

```python
df = pd.read_csv('data/musicData.csv')
```

```python
df.rename(columns={'Album Title' : '앨범명', 'Artist Name' : '가수명', 'Genre' : '장르', 'Date' : '날짜', 'Distribution' : '유통사', 'Rating' : '평점', 'Music Count' : '재생횟수'}, inplace=True)
df
```

1. 'Album Title'로 정렬 및 이름을 한글로 바꾸기

```python
df2 = df.copy()
df2 = df2.set_index('앨범명')
```

```python
df2
```

2. 평점 가장 높은 순부터 순차적으로 보이게 정렬한 뒤 재생 횟수를 기준으로 다시 정렬하기

```python
df3 = df2.copy()
df3.sort_values(by=['평점', '재생횟수'], ascending=False, inplace=True)
df3
```

3. 재생 횟수가 가장 높은 순으로 정렬하여, top5의 유통사를 index로 가수와 앨범명을 뽑아라

```python
df4 = df.copy()
df4 = df4.sort_values('재생횟수', ascending=False).set_index('유통사')
```

```python
df4[['앨범명', '가수명', '재생횟수']].head()
```

4. 장르가 '발라드' 인 것중에 가장 먼저 나온 노래의 가수와 재생횟수, 앨범명을 뽑고 index는 날짜로 지정해라

```python
df5 = df.copy()
```

```python
df5 = df5[df5['장르'] == '발라드'].sort_values('날짜').set_index('날짜')
```

```python
df5[['가수명', '재생횟수', '앨범명']].head(1)
```
