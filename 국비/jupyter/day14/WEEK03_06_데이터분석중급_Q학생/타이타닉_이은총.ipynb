## 15. 이름(Name) 컬럼에서 호칭(Title) 컬럼을 뽑아주세요.

```python
train[["Name"]]
```

```python
## re 함수 이용해서 풀기
import re
def get_title(name):
    pattern = r"[,.]"
    return re.split(pattern, name)[1].strip()

## 일반적인 함수 이용
def get_title2(name):
    return name.split(",")[1].split(".")[0].strip()
```

```python
get_title(train.loc[1, "Name"])
```

```python
## re 함수 적용

train["Title"] = train["Name"].apply(get_title)
train[["Name", "Title"]]
```

```python
## 일반 함수 적용

train["Title"] = train["Name"].apply(get_title2)
train[["Name", "Title"]]
```

```python
### 호칭별 생존률
```

```python
pd.pivot_table(train, index = "Title", values = "Survived")
```

### 16. 호칭(Title) 컬럼을 정리해주세요.

```python
## crosstab 활용
pd.crosstab(train["Survived(Status)"], train["Title"])
```

```python
### get_cleaned_title 함수 정의
def get_cleaned_title(name):
    if name == "Mr":
        return "Mr"
    elif name == "Mrs":
        return "Mrs"
    elif name == "Miss":
        return "Miss"
    elif name == "Master":
        return "Master"
    else:
        return "Other"
```

```python
train["Title"].info()
```

```python
train["Title(Status)"] = train["Title"].apply(get_cleaned_title)
train[["Name", "Title", "Title(Status)"]].tail(5)
```

```python
## 호칭별 생존률
pd.pivot_table(train, index = "Title(Status)", values = "Survived")
```
