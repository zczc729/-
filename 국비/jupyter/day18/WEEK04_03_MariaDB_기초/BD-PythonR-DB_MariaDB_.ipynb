```python
import os
import pandas as pd

# CSV 파일이 들어있는 디렉토리 경로
directory = './data/'

# 빈 리스트 생성
dfs = []

# 디렉토리 내 모든 CSV 파일 가져오기
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        print(f"Reading file: {filename}")
        
        # CSV 파일을 데이터프레임으로 읽기
        df = pd.read_csv(filepath, encoding='cp949')
        df = df.drop_duplicates()
        df['GU'] = filename.split("_")[1]
        df['YM'] = filename.split("_")[2].split(".")[0]
        df['TID'] = df.index + 1
        
        # 데이터프레임을 리스트에 추가
        dfs.append(df)
```

```python
df#.head(1)
```

```python
dfs
```

```python
# 리스트에 있는 모든 데이터프레임을 합치기
result = pd.concat(dfs, ignore_index=True)
result.info()
```

```python
#result[result.duplicated(keep=False)]
```

```python
# 결과 저장 및 확인
result.to_csv("result_all.csv", index=False, encoding='cp949')
```

```python
import pandas as pd
import pymysql
import numpy as np

# 데이터 프레임 생성
result = pd.read_csv("result_all.csv", index_col=None, encoding='cp949')
df = result
```

```python
# 데이터프레임의 NaN 값을 NULL로 대체
df = df.fillna("")
```

```python
##############################
# ChatGPT 활용 및 테이블 생성
##############################
# 판다스 데이터 프레임을 MariaDB에 저장하기 위한 테이블을 만들려고해. 
# 테이블 이름은 real_estate_seoul 으로 만들고
# 가급적 데이터가 짤리지 않게 DB 데이터타입을 넉넉하게 만들고 컬럼명도 영어로 변환해줘.
# 그리고, YM, GU, TID 컬럼이 기본키가 되게 해주고 형식은 문자열이 되게 해주고,
# 기본키는 테이블 맨앞 컬럼에 배치하고 컬럼명은 그대로 사용해줘.
# 연도 관련된 컬럼 건축년도, 등기일자, 해제사유발생일 컬럼과 지역코드는 문자열로 바꾸어줘
# 다음 데이터프레임 정보를 참고해서 만들려고 해.

# CREATE TABLE real_estate_seoul (
#     YM VARCHAR(10) NOT NULL,
#     GU VARCHAR(100) NOT NULL,
#     TID VARCHAR(50) NOT NULL,
#     transaction_amount VARCHAR(100),
#     transaction_type VARCHAR(100),
#     construction_year VARCHAR(100),
#     year VARCHAR(100),
#     dong VARCHAR(100),
#     registration_date VARCHAR(100),
#     seller VARCHAR(100),
#     buyer VARCHAR(100),
#     legal_dong VARCHAR(100),
#     apartment VARCHAR(100),
#     month INT,
#     day INT,
#     exclusive_area FLOAT,
#     agent_location VARCHAR(100),
#     jibun VARCHAR(100),
#     region_code VARCHAR(100),
#     floor INT,
#     cancellation_reason_date VARCHAR(100),
#     cancellation_status VARCHAR(100),
#     PRIMARY KEY (YM, GU, TID)
# );
```

```python
# pymysql을 사용하여 MariaDB에 연결
conn = pymysql.connect(host='127.0.0.1', user='user', password='user1234', db='dbmaria', charset='utf8')
# 커서 생성
cur = conn.cursor()
```

```python
# User 테이블 생성
# cur.execute('''CREATE TABLE IF NOT EXISTS user (
#                 nameUser VARCHAR(255),
#                 ageUser INTEGER,
#                 phoneUser VARCHAR(20)
#             )''')
```

```python
# # 데이터 프레임의 데이터를 데이터베이스에 삽입
for index, row in df.iterrows():
    cur.execute("""
        INSERT INTO real_estate_seoul (
            YM, GU, TID, 
            transaction_amount, 
            transaction_type, 
            construction_year, 
            year,
            dong, 
            registration_date, 
            seller, buyer, 
            legal_dong, 
            apartment, 
            month, 
            day,
            exclusive_area, 
            agent_location, 
            jibun, 
            region_code, 
            floor,
            cancellation_reason_date, 
            cancellation_status
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['YM'],
        row['GU'],
        row['TID'],
        row['거래금액'],
        row['거래유형'],
        row['건축년도'],
        row['년'],
        row['동'],
        row['등기일자'],
        row['매도자'],
        row['매수자'],
        row['법정동'],
        row['아파트'],
        row['월'],
        row['일'],
        row['전용면적'],
        row['중개사소재지'],
        row['지번'],
        row['지역코드'],
        row['층'],
        row['해제사유발생일'],
        row['해제여부']
    ))
```

```python
# 변경사항 커밋
conn.commit()

# 연결 종료
conn.close()
```
